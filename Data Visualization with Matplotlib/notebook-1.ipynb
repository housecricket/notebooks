```python
t = np.arange(0,5,0.1)

y1 = np.sin(2*np.pi*t)
y2 = np.sin(2*np.pi*t)

plt.subplot(211)
plt.plot(t,y1,'b-.')
plt.subplot(212)
plt.plot(t,y2,'r--')

plt.show()
```

```python
t = np.arange(0.,1.,0.05)

y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)

plt.subplot(121)
plt.plot(t,y1,'b-.')
plt.subplot(122)
plt.plot(t,y2,'r--')

plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = ['Viettel', 'VNPT', 'Mobiphone']
revenue = (37600000, 6445000, 6045000)


x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, revenue, color='green')
plt.xlabel("Telco")
plt.ylabel("VND")
plt.title("Telecom service revenues in Vietnam")

plt.xticks(x_pos, x)

plt.show()
```

```python
N = 3

revenue = (37600000, 6445000, 6045000)
tax = (37000000, 4476000, 5403000)

ind = np.arange(N) 
width = 0.35       
plt.bar(ind, revenue, width, label='Revenue')
plt.bar(ind + width, tax, width,
    label='Tax')

plt.ylabel('VND')
plt.title('Telecom service revenues in Vietnam')

plt.xticks(ind + width / 2, ('Viettel', 'VNPT', 'Mobiphone'))
plt.legend(loc='best')
plt.show()
```

```python
telcos = ['Viettel', 'VNPT', 'Mobiphone']

revenue = (37600000, 6445000, 6045000)
tax = (37000000, 4476000, 5403000)

ind = [x for x, _ in enumerate(telcos)]

plt.bar(ind, revenue, width=0.8, label='revenue', color='gold', bottom=tax)
plt.bar(ind, tax, width=0.8, label='tax', color='silver')

plt.xticks(ind, telcos)
plt.ylabel("VND")
plt.xlabel("Telco")
plt.legend(loc="upper right")
plt.title("Telecom service revenues in Vietnam")

plt.show()
```
